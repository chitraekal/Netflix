```python
mode_country_by_director = df.groupby('director')['country'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else 'Not Mentioned')

# Fill NaN values using the mode
df['country'] = df['country'].fillna(df['director'].map(mode_country_by_director))
```

```python
valid_dates = df.date_added[(df.date_added.dt.year >= df.release_year) & (df.date_added.dt.year==df.release_year+4)]
random_date_per_year = df.groupby('release_year')['date_added'].apply(lambda x: valid_dates.sample(1).iloc[0])
print(random_date_per_year)
```

```python
df['date_added'] = df['date_added'].fillna(df['release_year'].map(random_date_per_year))
```
